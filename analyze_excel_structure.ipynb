## Section 8: Reorganize Excel File to Target Structure

```python
from openpyxl import Workbook
from openpyxl.styles import Font, PatternFill, Alignment
from openpyxl.utils import get_column_letter
import datetime

# Define the target column structure (29 columns A-AC)
target_columns = [
    'S.No',
    'Provision Month',
    'ATM BNA ID',
    'Docket No',
    'Bank Name',
    'From Location',
    'From State',
    'To Location',
    'To State',
    'Business Group',
    'Mode of Bill',
    'Type of Movement',
    'Assets Service Description',
    'Total Cost',
    'Hold',
    'Deduction',
    'Final Amount',
    'Vendor',
    'Assets Delivery Pending',
    'Reason for Additional Charges',
    'Pick Up Date',
    'Status',
    'Date',
    'Vendor Name',
    'Freight Category',
    'Project',
    'Invoice No',
    'Billing Month',
    'Billing'
]

print("="*100)
print("REORGANIZING EXCEL FILE TO TARGET STRUCTURE")
print("="*100)

# Ensure we have the right number of columns
assert len(target_columns) == 29, f"Expected 29 columns but got {len(target_columns)}"

print(f"\nâœ“ Target structure has {len(target_columns)} columns (A-AC)")
print("\nTarget columns:")
for idx, col in enumerate(target_columns, 1):
    excel_col = get_column_letter(idx)
    print(f"  {excel_col}: {col}")

# Load current data
file_path = r"d:\internship\assetflow-hub\NCR File.xlsx"
current_df = pd.read_excel(file_path)

print(f"\n\nCURRENT FILE ANALYSIS:")
print(f"  File Path: {file_path}")
print(f"  Current Rows: {len(current_df)}")
print(f"  Current Columns: {len(current_df.columns)}")

# Print current columns for mapping reference
print(f"\nCurrent columns in source file:")
for idx, col in enumerate(current_df.columns, 1):
    print(f"  {idx:2d}. {col}")

# Create mapping between current columns and target columns
# This is a flexible mapping that tries to match column names
print(f"\n\nCOLUMN MAPPING STRATEGY:")
print("-"*100)

# Create new dataframe with target structure
new_df = pd.DataFrame(columns=target_columns)

# Define mapping from current columns to target columns
# This maps based on column name similarity and logical relationships
column_mapping = {}

current_cols_lower = {col.lower().strip(): col for col in current_df.columns}

# Manual mapping based on the expected structure
mapping_rules = {
    'S.No': None,  # Will be generated
    'Provision Month': ['provision month', 'month', 'prov.month'],
    'ATM BNA ID': ['atm bna id', 'atm id', 'bna id'],
    'Docket No': ['docket no', 'docket', 'docket number'],
    'Bank Name': ['bank name', 'bank'],
    'From Location': ['from location', 'source location', 'from'],
    'From State': ['from state', 'source state'],
    'To Location': ['to location', 'destination location', 'to'],
    'To State': ['to state', 'destination state'],
    'Business Group': ['business group', 'business', 'bg'],
    'Mode of Bill': ['mode of bill', 'mode', 'billing mode'],
    'Type of Movement': ['type of movement', 'movement', 'movement type'],
    'Assets Service Description': ['assets service description', 'description', 'service description'],
    'Total Cost': ['total cost', 'cost', 'total'],
    'Hold': ['hold', 'hold amount'],
    'Deduction': ['deduction', 'deduct'],
    'Final Amount': ['final amount', 'amount', 'final'],
    'Vendor': ['vendor', 'vendor id'],
    'Assets Delivery Pending': ['assets delivery pending', 'delivery pending', 'pending'],
    'Reason for Additional Charges': ['reason for additional charges', 'reason', 'charges'],
    'Pick Up Date': ['pick up date', 'pickup date', 'pickup'],
    'Status': ['status'],
    'Date': ['date', 'transaction date'],
    'Vendor Name': ['vendor name', 'vendor_name'],
    'Freight Category': ['freight category', 'freight', 'category'],
    'Project': ['project', 'project name'],
    'Invoice No': ['invoice no', 'invoice', 'invoice number'],
    'Billing Month': ['billing month', 'billing'],
    'Billing': ['billing code', 'bill code']
}

# Find matches for each target column
matches_found = {}
unmapped_source_cols = set(current_cols_lower.keys())

for target_col, search_terms in mapping_rules.items():
    if search_terms is None:  # S.No case
        continue
    
    matched_col = None
    for search_term in search_terms:
        if search_term in current_cols_lower:
            matched_col = current_cols_lower[search_term]
            break
    
    if matched_col:
        matches_found[target_col] = matched_col
        unmapped_source_cols.discard(matched_col.lower().strip())
    else:
        matches_found[target_col] = None

print("\nMapping Results:")
for target_col, source_col in matches_found.items():
    if source_col:
        print(f"  âœ“ {target_col:35s} <- {source_col}")
    else:
        print(f"  âœ— {target_col:35s} <- [NOT FOUND]")

if unmapped_source_cols:
    print(f"\nUnmapped source columns (will not be transferred):")
    for col in sorted(unmapped_source_cols):
        print(f"  â€¢ {col}")

# Build the new dataframe
print(f"\n\nBUILDING NEW DATAFRAME:")
print("-"*100)

# Initialize with empty rows
new_df = pd.DataFrame(index=range(len(current_df)), columns=target_columns)

# Add S.No column (row numbers)
new_df['S.No'] = range(1, len(current_df) + 1)

# Map data from current dataframe to new dataframe
for target_col, source_col in matches_found.items():
    if source_col and source_col in current_df.columns:
        new_df[target_col] = current_df[source_col].values
    else:
        new_df[target_col] = pd.NA  # Leave empty if not found

# Ensure numeric columns are properly typed
numeric_columns = ['Total Cost', 'Hold', 'Deduction', 'Final Amount']
for col in numeric_columns:
    if col in new_df.columns:
        new_df[col] = pd.to_numeric(new_df[col], errors='coerce')

print(f"\nâœ“ Created new dataframe with {len(new_df)} rows and {len(target_columns)} columns")

# Display statistics
print(f"\n\nDATA SUMMARY:")
print("-"*100)
print(f"Total rows processed: {len(new_df)}")

# Count missing data
missing_data = {}
for col in target_columns:
    missing_count = new_df[col].isna().sum()
    if missing_count > 0:
        missing_data[col] = missing_count

if missing_data:
    print(f"\nColumns with unmapped/missing data:")
    for col, count in sorted(missing_data.items()):
        percent = (count / len(new_df)) * 100
        print(f"  â€¢ {col:35s}: {count:5d} rows ({percent:5.1f}%)")
else:
    print(f"\nâœ“ All columns have been populated with data!")

# Save the reorganized file
print(f"\n\nSAVING REORGANIZED FILE:")
print("-"*100)

try:
    # Create a new workbook using openpyxl for better formatting
    wb = Workbook()
    ws = wb.active
    ws.title = "NCR Data"
    
    # Write headers with formatting
    header_fill = PatternFill(start_color="4472C4", end_color="4472C4", fill_type="solid")
    header_font = Font(bold=True, color="FFFFFF")
    
    for col_idx, col_name in enumerate(target_columns, 1):
        cell = ws.cell(row=1, column=col_idx)
        cell.value = col_name
        cell.fill = header_fill
        cell.font = header_font
        cell.alignment = Alignment(horizontal='center', vertical='center')
    
    # Write data rows
    for row_idx, row_data in enumerate(new_df.values, 2):
        for col_idx, value in enumerate(row_data, 1):
            cell = ws.cell(row=row_idx, column=col_idx)
            # Handle different data types
            if pd.isna(value):
                cell.value = None
            elif isinstance(value, datetime.datetime):
                cell.value = value
            else:
                cell.value = value
    
    # Auto-adjust column widths
    for col_idx, col_name in enumerate(target_columns, 1):
        max_length = len(col_name)
        ws.column_dimensions[get_column_letter(col_idx)].width = min(max_length + 2, 30)
    
    # Save the workbook
    wb.save(file_path)
    
    print(f"âœ“ File saved successfully: {file_path}")
    print(f"  - Output Rows: {len(new_df)}")
    print(f"  - Output Columns: {len(target_columns)}")
    print(f"  - File size: {os.path.getsize(file_path) / 1024:.2f} KB")
    
except Exception as e:
    print(f"âœ— Error saving file: {str(e)}")
    print(f"  Attempting alternative save method...")
    new_df.to_excel(file_path, index=False)
    print(f"âœ“ File saved using pandas method")

print("\n" + "="*100)
print("REORGANIZATION COMPLETE")
print("="*100)

# Final summary
print(f"\nðŸ“Š FINAL REPORT:")
print(f"   âœ“ Rows processed: {len(new_df)}")
print(f"   âœ“ Columns in output: {len(target_columns)}")
print(f"   âœ“ Output structure: Columns A through AC")
print(f"   âœ“ File saved: {file_path}")

# Show which columns had data mapped
mapped_count = sum(1 for col in target_columns if col != 'S.No' and matches_found.get(col) is not None)
print(f"\n   Data mapping:")
print(f"   âœ“ Columns with data mapped: {mapped_count}/{len(target_columns)-1}")

# Show columns with missing data
if missing_data:
    print(f"   âš  Columns with unmapped data (empty): {len(missing_data)}")
else:
    print(f"   âœ“ All columns have data!")

print(f"\n   File Status: READY FOR USE")
```
